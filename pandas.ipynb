### Slicing and Dicing the <span style='color:red'>DataFrame</span> and <span style='color:red'>Series</span>

In this section, we will expand on topics covered in the previous sections on this page. One of the most important concepts to learn about Pandas is that it allows you to access anything by its associated label, regardless of data organization structure.


#### Indexing a <span style='color:red'>Series

As a review of previous examples, weâ€™ll start our next example by pulling a <span style='color:red'>Series</span> out of our <span style='color:red'>DataFrame</span> using its column label.

```python
nino34_series = df["Nino34"]
nino34_series
```

You can use syntax similar to that of NumPy <span style='color:red'>ndarrays</span> to index, select, and subset with Pandas <span style='color:red'>Series</span>, as shown in this example:

```python
nino34_series[3]
```

You can also use labels alongside Python dictionary syntax to perform the same operations:

```python
nino34_series["1982-04-01"]
```

You can probably figure out some ways to extend these indexing methods, as shown in the following examples:

```python
nino34_series[0:12]
```

However, there are many more ways to index a <span style='color:red'>Series</span>. The following example shows a powerful and useful indexing method:

```python
nino34_series["1982-01-01":"1982-12-01"]
```

This is an example of label-based slicing. With label-based slicing, Pandas will automatically find a range of values based on the labels you specify.

As opposed to index-based slices, label-based slices are inclusive of the final value.

If you already have some knowledge of xarray, you will quite likely know how to create <span style='color:red'>slice</span> objects by hand. This can also be used in pandas, as shown below. If you are completely unfamiliar with xarray, it will be covered on a later Pythia tutorial page

```python
nino34_series[slice("1982-01-01", "1982-12-01")]
```

### Using <span style='color:red'>.iloc</span> and <span style='color:red'>.loc</span> to index

In this section, we introduce ways to access data that are preferred by Pandas over the methods listed above. When accessing by label, it is preferred to use the <span style='color:red'>.loc</span> method, and when accessing by index, the <span style='color:red'>.iloc</span> method is preferred. These methods behave similarly to the notation introduced above, but provide more speed, security, and rigor in your value selection. Using these methods can also help you avoid chained assignment warnings generated by pandas.

```python
nino34_series.iloc[3]
```

```python
nino34_series.iloc[0:12]
```

```python
nino34_series.loc["1982-04-01"]
```

```python
nino34_series.loc["1982-01-01":"1982-12-01"]
```

### Extending to the <span style='color:red'>DataFrame</span>

These subsetting capabilities can also be used in a full <span style='color:red'>DataFrame</span>; however, if you use the same syntax, there are issues, as shown below:

```python
df["1982-01-01"]
```

### <b><span style='color:red'>Danger</span></b>

Attempting to use Series subsetting with a <span style='color:red'>DataFrame</span> can crash your program. A proper way to subset a <span style='color:red'>DataFrame</span> is shown below.


When indexing a <span style='color:red'>DataFrame</span>, pandas will not assume as readily the intention of your code. In this case, using a row label by itself will not work; <b>with</b> <span style='color:red'>DataFrames</span>, <b>labels are used for identifying columns</b>.

```python
df['Nino34']
```

As shown below, you also cannot subset columns in a <span style='color:red'>DataFrame</span> using integer indices:

```python
df[0]
```

From earlier examples, we know that we can use an index or label with a <span style='color:red'>DataFrame</span> to pull out a column as a <span style='color:red'>Series</span>, and we know that we can use an index or label with a <span style='color:red'>Series</span> to pull out a single value. Therefore, by chaining brackets, we can pull any individual data value out of the <span style='color:red'>DataFrame</span>.

```python
df["Nino34"]["1982-04-01"]
```

```python
df["Nino34"][3]
```

However, subsetting data using this chained-bracket technique is not preferred by Pandas. As described above, Pandas prefers us to use the <span style='color:red'>.loc</span> and <span style='color:red'>.iloc</span> methods for subsetting. In addition, these methods provide a clearer, more efficient way to extract specific data from a <span style='color:red'>DataFrame</span>, as illustrated below:

```python
df.loc["1982-04-01", "Nino34"]
```

### <b>Info</b>

When using this syntax to pull individual data values from a DataFrame, make sure to list the row first, and then the column.

The <span style='color:red'>.loc</span> and <span style='color:red'>.iloc</span> methods also allow us to pull entire rows out of a <span style='color:red'>DataFrame</span>, as shown in these examples:

```python
df.loc["1982-04-01"]
```

```python
df.loc["1982-01-01":"1982-12-01"]
```

```python
df.iloc[0:12]
```

In the next example, we illustrate how you can use slices of rows and lists of columns to create a smaller <span style='color:red'>DataFrame</span> out of an existing <span style='color:red'>DataFrame</span>:

```python
df.loc[
    "1982-01-01":"1982-12-01",  # slice of rows
    ["Nino12", "Nino3", "Nino4", "Nino34"],  # list of columns
]
```

### Quick Plots of your Data

A good way to explore your data is by making a simple plot. Pandas contains its own <span style='color:red'>plot</span> method; this allows us to plot Pandas series without needing <span style='color:red'>matplotlib</span>. In this example, we plot the <span style='color:red'>Nino34</span> series of our <span style='color:red'>df DataFrame</span> in this way:

```python
df.Nino34.plot()
```

Before, we called <span style='color:red'>.plot()</span>, which generated a single line plot. Line plots can be helpful for understanding some types of data, but there are other types of data that can be better understood with different plot types. For example, if your data values form a distribution, you can better understand them using a histogram plot.

The code for plotting histogram data differs in two ways from the code above for the line plot. First, two series are being used from the <span style='color:red'>DataFrame</span> instead of one. Second, after calling the <span style='color:red'>plot</span> method, we call an additional method called <span style='color:red'>hist</span>, which converts the plot into a histogram.

```python
df[['Nino12', 'Nino34']].plot.hist()
```

The histogram plot helped us better understand our data; there are clear differences in the distributions. To even better understand this type of data, it may also be helpful to create a box plot. This can be done using the same line of code, with one change: we call the <span style='color:red'>box</span> method instead of <span style='color:red'>hist</span>.

```python
df[['Nino12', 'Nino34']].plot.box()
```

Just like the histogram plot, this box plot indicates a clear difference in the distributions. Using multiple types of plot in this way can be useful for verifying large datasets. The pandas plotting methods are capable of creating many different types of plots. To see how to use the plotting methods to generate each type of plot, please review the pandas plot documentation (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html).

#### Customize your plot

### Subsetting Using the Datetime Column

Slicing is a useful technique for subsetting a <span style='color:red'>DataFrame</span>, but there are also other options that can be equally useful. In this section, some of these additional techniques are covered.

If your <span style='color:red'>DataFrame</span> uses datetime values for indices, you can select data from only one month using <span style='color:red'>df.index.month</span>. In this example, we specify the number 1, which only selects data from January.

```python
# Uses the datetime column
df[df.index.month == 1]
```

This example shows how to create a new column containing the month portion of the datetime index for each data row. The value returned by <span style='color:red'>df.index.month</span> is used to obtain the data for this new column:

```python
df['month'] = df.index.month
```

This next example illustrates how to use the new month column to calculate average monthly values over the other data columns. First, we use the <span style='color:red'>groupby</span> method to group the other columns by the month. Second, we take the average (mean) to obtain the monthly averages. Finally, we plot the resulting data as a line plot by simply calling <span style='color:red'>plot()</span>.

```python
df.groupby('month').mean().plot();
```

### Resampling

In these examples, we illustrate a process known as resampling. Using resampling, you can change the frequency of index data values, reducing so-called â€˜noiseâ€™ in a data plot. This is especially useful when working with timeseries data; plots can be equally effective with resampled data in these cases. The resampling performed in these examples converts monthly values to yearly averages. This is performed by passing the value â€˜1Yâ€™ to the <span style='color:red'>resample</span> method.

```python
df.Nino34.plot();
```

```python
df.Nino34.resample('1Y').mean().plot();
```

### Applying operations to a DataFrame

One of the most commonly used features in Pandas is the performing of calculations to multiple data values in a <span style='color:red'>DataFrame</span> simultaneously. Letâ€™s first look at a familiar concept: a function that converts single values. The following example uses such a function to convert temperature values from degrees Celsius to Kelvin.

```python
def convert_degc_to_kelvin(temperature_degc):
    """ Converts from degrees celcius to Kelvin """
    return temperature_degc + 273.15
```

```python
# Convert to a single value
convert_degc_to_kelvin(0)
```

The following examples instead illustrate a new concept: using such <span style='color:red'>functions</span> with <span style='color:red'>DataFrames</span> and <span style='color:red'>Series</span>. For the first example, we start by creating a <span style='color:red'>Series</span>; in order to do so, we subset the <span style='color:red'>DataFrame</span> by the <span style='color:red'>Nino34</span> column. This has already been done earlier in this page; we do not need to create this <span style='color:red'>Series</span> again. We are using this particular <span style='color:red'>Series</span> for a reason: the data values are in degrees Celsius.

```python
nino34_series
```

Here, we look at a portion of an existing <span style='color:red'>DataFrame</span> column. Notice that this column portion is a Pandas <span style='color:red'>Series</span>.

```python
type(df.Nino12[0:10])
```

As shown in the following example, each Pandas <span style='color:red'>Series</span> contains a representation of its data in numpy format. Therefore, it is possible to convert a Pandas <span style='color:red'>Series</span> into a numpy array; this is done using the <span style='color:red'>.values</span> method:

```python
type(df.Nino12.values[0:10])
```

This example illustrates how to use the temperature-conversion function defined above on a <span style='color:red'>Series</span> object. Just as calling the function with a single value returns a single value, calling the function on a <span style='color:red'>Series</span> object returns another <span style='color:red'>Series</span> object. The function performs the temperature conversion on each data value in the <span style='color:red'>Series</span>, and returns a <span style='color:red'>Series</span> with all values converted.

```python
convert_degc_to_kelvin(nino34_series)
```

If we call the <span style='color:red'>.values</span> method on the <span style='color:red'>Series</span> passed to the function, the Series is converted to a numpy array, as described above. The function then converts each value in the numpy array, and returns a new numpy array with all values sorted.

```python
convert_degc_to_kelvin(nino34_series.values) # The nino_series is converted to a numpy array
```

As described above, when our temperature-conversion function accepts a <span style='color:red'>Series</span> as an argument, it returns a <span style='color:red'>Series</span>. We can directly assign this returned <span style='color:red'>Series</span> to a new column in our <span style='color:red'>DataFrame</span>, as shown below:

```python
df['Nino34_degK'] = convert_degc_to_kelvin(nino34_series)
```

```python
df['Nino34_degK']
```
