### Basic Line Plots
* Let’s create a Figure whose dimensions, if printed out on hardcopy, would be 10 inches wide and 6 inches long (assuming a landscape orientation).

* After that, we call the Axes object’s plot method, using the times array for the data along the x-axis (i.e., the independent values), and the temps array for the data along the y-axis (i.e., the dependent values).

```python
# Create a figure with dimensions - width 10inches, height 6inches
fig = plt.figure(figsize=(10, 6))

# Ask, out of a 1x1 grid of plots, the first axes.
ax = fig.add_subplot(1, 1, 1)

# Plot times as x-variable and temperature as y-variable
ax.plot(times, temps)
```

#### Adding labels and a grid

Here, we call <b>plot</b> more than once, in order to plot multiple series of temperature data on the same plot. We also specify the <b>label</b> keyword argument to the <b>plot</b> method to allow Matplotlib to automatically create legend labels. These legend labels are added via a call to the <b>legend</b> method. By utilizing the <b>grid()</b> method, we can also add gridlines to our plot.

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1)

# Plot two series of data
# The label argument is used when generating a legend.
ax.plot(times, temps, label='Temperature (surface)')
ax.plot(times, temps_1000, label='Temperature (1000 mb)')

# Add labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Temperature')
ax.set_title('Temperature Forecast')

# Add gridlines
ax.grid(True)

# Add a lagend to the upper left corner of the plot
ax.legend(loc='best')
```

#### Customizing colors

* For example, the <b>color</b> attribute can accept a wide array of color options, including keywords (named colors) like <b>red</b> or <b>blue</b>, or HTML color codes. 
* Here, we use some different shades of red taken from the Tableau colorset in Matplotlib, by using the <b>tab:red</b> option for the color attribute.


```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1)

# Specify how our lines should look
ax.plot(times, temps, color='tab:red', label='Temperature (surface)')
ax.plot(times, temps_1000, color='tab:red', linestyle='--', label='Temperature (isobaric level)')

# Set the labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Temperature')
ax.set_title('Temperature Forecast')

# Add the grid
ax.grid(True)

# Add a legend 
ax.legend(loc='best')

```

#### Subplots

The term “subplots” refers to working with multiple plots, or panels, in a figure.

Here, we create yet another set of test data, in this case dew-point data, to be used in later examples:

```python
dewpoint = 0.9 * temps
dewpoint_1000 = 0.9 * temps_1000
```

Now, we can use subplots to plot this new data alongside the temperature data.

```python
fig = plt.figure(figsize=(10, 6))

# Create a plot for temperature
ax = fig.add_subplot(1, 2, 1)
ax.plot(times, temps, color='tab:red')

# Create a plot for dewpoint
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(times, dewpoint, color='tab:green')
```

You can also call plot.subplots() with the keyword arguments <b>nrows</b> (number of rows) and <b>ncols</b> (number of columns). This initializes a new Axes object, called ax, with the specified number of rows and columns. This object also contains a 1-D list of subplots, with a size equal to <b>nrows</b> X <b>ncols</b>.

You can index this list, using ax[0].plot(), for example, to decide which subplot you’re plotting to. Here is some example code for this technique:

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 6))

ax[0].plot(times, temps, color='tab:red')
ax[1].plot(times, dewpoint, color='tab:green')
```

#### Adding titles to each subplot
We can add titles to these plots too; notice that these subplots are titled separately, by calling ax.set_title after plotting each subplot:

```python
fig = plt.figure(figsize=(10, 6))

# Create a plot for temperature
ax = fig.add_subplot(1, 2, 1)
ax.plot(times, temps, color='tab:red')
ax.set_title('Temperature')

# Create a plot for dewpoint
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(times, dewpoint, color='tab:green')
ax2.set_title('Dewpoint')
```

#### Using <b>ax.set_xlim</b> and <b>ax.set_ylim</b> to control the plot boundaries
It is common when plotting data to set the extent (boundaries) of plots, which can be performed by calling <b>.set_xlim</b> and <b>.set_ylim</b> on the Axes object containing the plot or subplot(s):

```python
fig = plt.figure(figsize=(10, 6))

# Create a plot for temperature
ax = fig.add_subplot(1, 2, 1)
ax.plot(times, temps, color='tab:red')
ax.set_title('Temperature')
ax.set_xlim(110, 130)
ax.set_ylim(290, 315)

# Create a plot for dewpoint
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(times, dewpoint, color='tab:green')
ax2.set_title('Dewpoint')
ax2.set_xlim(110, 130)
```

#### Using sharex and sharey to share plot limits
You may want to have both subplots share the same x/y axis limits. When setting up a new Axes object through a method like <b>add_subplot</b>, specify the keyword arguments <b>sharex=ax</b> and <b>sharey=ax</b>, where <b>ax</b> is the <b>Axes</b> object with which to share axis limits.

Let’s take a look at an example:

```python
fig = plt.figure(figsize=(10, 6))

# Create a plot for temperature
ax = fig.add_subplot(1, 2, 1)
ax.plot(times, temps, color='tab:red')
ax.set_title('Temperature')
ax.set_ylim(260, 320)

# Create a plot for dewpoint
ax2 = fig.add_subplot(1, 2, 2, sharex=ax, sharey=ax)
ax2.plot(times, dewpoint, color='tab:green')
ax2.set_title('Dewpoint');
```

#### Putting this all together

```python
fig = plt.figure(figsize=(10, 6))

# For the first plot
ax = fig.add_subplot(1, 2, 1)

# Specify how our lines should look (and simply generate the first plot)
ax.plot(times, temps, color='tab:red', label='Temperature (surface)')
ax.plot(times, temps_1000, color='tab:red', linestyle=':', label='Temperature (isobaric level)')

# Add labels, grid, and legend
ax.set_xlabel('Time')
ax.set_ylabel('Temperature')
ax.set_title('Temperature Forecast')
ax.grid(True)
ax.legend(loc='best')
ax.set_ylim(257, 312)
ax.set_xlim(95, 162)



# For the second plot, let it share x and y limit with the first plot
ax2 = fig.add_subplot(1, 2, 2, sharex=ax, sharey=ax)

# Specify how our lines should look (for the second plot)
ax2.plot(times, dewpoint, color='tab:green', label='Dewpoint (surface)')
ax2.plot(times, dewpoint_1000, color='tab:green', linestyle=':', marker='o', label='Dewpoint (isobaric level)')


# Add labels, grid, and legend
ax2.set_xlabel('Time')
ax2.set_ylabel('Dewpoint')
ax2.set_title('Dewpoint Forecast')
ax2.grid(True)
ax2.legend(loc='best')
```

## Scatterplot

Some data cannot be plotted accurately as a line plot. Another type of plot that is popular in science is the marker plot, more commonly known as a scatter plot. A simple scatter plot can be created by setting the <b>linestyle</b> to <b>None</b>, and specifying a marker type, size, color, etc., like this:

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1, 1, 1)

# Specify no line with circle markers
ax.plot(temps, temps_1000, linestyle='None', marker='o', markersize=5)

# Add labels, grid, and legend
ax.set_xlabel('Temperature (surface)')
ax.set_ylabel('Temperature (1000 hPa)')
ax.set_title('Temperature Cross Plot')
ax.grid(True)
ax.legend(loc='best') # No legend can be added for this
```

You can also use the <b>scatter</b> method, which is slower, but will give you more control, such as being able to color the points individually based upon a third variable.

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1)

# Use the scatter method here
ax.scatter(temps, temps_1000)

# Add labels, grid, and legend
ax.set_xlabel('Temperature (surface)')
ax.set_ylabel('Temperature (1000 hPa)')
ax.set_title('Temperature Cross Plot')
ax.grid(True)
```

#### Going further

* Let’s put together the following:

* Beginning with our code above, add the c keyword argument to the scatter call; in this case, to color the points by the difference between the temperature at the surface and the temperature at 1000 hPa.

* Add a 1:1 line to the plot (slope of 1, intercept of zero). Use a black dashed line.

* Change the colormap to one more suited for a temperature-difference plot.

* Add a colorbar to the plot (have a look at the Matplotlib documentation for help).

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(1, 1, 1)

# Add the slope of the plot using a black-dashed line
ax.plot([285, 320], [285, 320], color='black', linestyle='--')

# Add the scatterplot, change the colormap to a temperature-difference plot(c)
s= ax.scatter(temps, temps_1000, c=(temps - temps_1000), cmap='bwr', vmin=-5, vmax=5)

# Add appropriate legend
fig.colorbar(s)

# Add labels, grid, and legend
ax.set_xlabel('Temperature (surface)')
ax.set_ylabel('Temperature (1000 hPa)')
ax.set_title('Temperature Cross Plot')
ax.grid(True)



```

## Displaying Images

<b>imshow</b> displays the values in an array as colored pixels, similar to a heat map.

Here, we declare some fake data in a bivariate normal distribution, to illustrate the <b>imshow</b> method:

```python
x = y = np.arange(-3.0, 3.0, 0.025)
X, Y = np.meshgrid(x, y)
Z1 = np.exp(-(X**2) - Y**2)
Z2 = np.exp(-((X - 1) ** 2) - (Y - 1) ** 2)
Z = (Z1 - Z2) * 2
```

We can now pass this fake data to imshow to create a heat map of the distribution:

```python
fig, ax = plt.subplots()
im = ax.imshow(
    Z, interpolation='bilinear', cmap='RdYlGn', origin='lower', extent=[-3, 3, -3, 3]
)
```

## Contour and Filled Contour Plots
* <b>contour</b> creates countours around data
* <b>contourf</b> creates filled contours around data

Let’s start with the <b>contour</b> method, which, as just mentioned, creates contours around data:

```python
fig, ax = plt.subplots()
ax.contour(X, Y, Z)
```

After creating contours, we can label the lines using the <b>clabel</b> method, like this:

```python
fig, ax = plt.subplots()
c = ax.contour(X, Y, Z, levels=np.arange(-2, 2, 0.25))
ax.clabel(c)
```

As described above, the contourf (contour fill) method creates filled contours around data, like this:

```python
fig, ax = plt.subplots()
c = ax.contourf(X, Y, Z);
```

As a final example, let’s create a heatmap figure with contours using the contour and imshow methods. First, we use imshow to create the heatmap, specifying a colormap using the cmap keyword argument. We then call contour, specifying black contours and an interval of 0.5. Here is the example code, and resulting figure:

```python
fig, ax = plt.subplots()
im = ax.imshow(
    Z, interpolation='bilinear', cmap='PiYG', origin='lower', extent=[-3, 3, -3, 3]
)
c = ax.contour(X, Y, Z, levels=np.arange(-2, 2, 0.5), colors='black')
ax.clabel(c);
```
